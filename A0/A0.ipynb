## Question 3

```julia
using Plots
```

```julia
function draw_hyperplane_2D(a, b)
    pyplot()
    x=range(-3, stop=3, length=100)
    g(x) = (dot(a, b) - a[1]*x)/a[2]
    p= plot(x, g, 
        title="2d hyperplane",
        label="$(a)'(x-$(b))",
        xlabel="x_1", ylabel="x_2",
        xlims=(-3, 3), ylims=(-3, 3),
        aspect_ratio=1)
    b_x1 = range(0, stop=b[1], length=100)
    b_x2 = range(0, stop=b[2], length=100)
    plot!(b_x1, b_x2, label="b")
    a_x1 = range(b[1], stop=a[1]+b[1], length=100)
    a_x2 = range(b[2], stop=a[2]+b[2], length=100)
    plot!(a_x1, a_x2, label="n")
    ylabel!(p, "x_2", label_position="left")
end
```

### Homework example

```julia
a = [-1; -1]
b = [2; 1]
draw_hyperplane_2D(a, b)
```

## Question 4 _Hyperplane in 3D World_


```julia
function draw_hyperplane_3D(a, b, camera=(60, 45))
    pyplot()
    x=range(-2, stop=2, length=100)
    y=range(-2, stop=2, length=100)
    g(x, y) = (dot(a, b) - dot(a[1: 2], [x; y]))/a[3]
    p= plot(x, y, g, 
        title="Hyperplane in 3D",
        label="$(a)'(x-$(b))=0",
        xlabel="x_1", ylabel="x_2", zlabel="x_3",
        xlims= (-2, 2), ylims=(-2, 2), zlims=(-2, 2),
        st=:surface, camera=camera)
    
    # plot the normal vector, since b is on the plane plot the vector b to b+a
    a_x1 = range(b[1], stop=(b[1]+a[1]), length=100)
    a_x2 = range(b[2], stop=(b[2]+a[2]), length=100)
    a_x3 = range(b[3], stop=(b[3]+a[3]), length=100)
    plot!(a_x1, a_x2, a_x3, label="a")
end
```

### Homework example

```julia
a = [1; 1; 1]
b = [2; 1; 0]
draw_hyperplane_3D(a, b, (30, 45))
```

```julia
# midterm
a = [1; 1; 0]
b = [0; 0; 0]
draw_hyperplane_3D(a, b, (30,45))
```
